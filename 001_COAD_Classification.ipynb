```python
# Define the neural network model
class SimpleClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(SimpleClassifier, self).__init__()
        self.f1 = nn.Linear(input_size, hidden_size)
        self.b1 = nn.BatchNorm1d(hidden_size)
        self.d1 = nn.Dropout(0.3)
        self.a1 = nn.ReLU(inplace=True)
        
        self.f2 = nn.Linear(hidden_size, hidden_size)
        self.b2 = nn.BatchNorm1d(hidden_size)
        self.d2 = nn.Dropout(0.2)
        self.a2 = nn.LeakyReLU(inplace=True)
        
        
        self.f3 = nn.Linear(hidden_size, hidden_size)
        self.b3 = nn.BatchNorm1d(hidden_size)
        self.d3 = nn.Dropout(0.1)
        self.a3 = nn.LeakyReLU(inplace=True)
        
        
        self.f4 = nn.Linear(hidden_size, hidden_size)
        self.b4 = nn.BatchNorm1d(hidden_size)
        self.d4 = nn.Dropout(0.1)
        self.a4 = nn.ReLU(inplace=True)
        
        
        self.final = nn.Linear(hidden_size, num_classes)
        self.sm = nn.LogSoftmax(dim=1)
    
    def forward(self, x):
        x1 = self.a1(self.d1(self.b1(self.f1(x))))
        x2 = self.a2(self.d2(self.b2(self.f2(x1)+x1)))
        x3 = self.a3(self.d3(self.b3(self.f3(x2))))
        x4 = self.a4(self.d4(self.b4(self.f4(x3)+x3)))
        
        x_final = self.final(x4)
        return self.sm(x_final)
```

```python
test_acc = []
conf_mat_sum_nn = []
for i in range(100):
    seed = i
    
    # Convert your dataset to PyTorch tensors and move to GPU if available
    X_sme_torch = torch.tensor(X_sme, dtype=torch.float32)
    y_sme_torch = torch.tensor(y_sme, dtype=torch.long)  # Assuming y contains integer class labels

    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        X_sme_torch,
        y_sme_torch,
        stratify=y_sme_torch,
        test_size=0.01,
        random_state=seed
    )

    # Instantiate the model and move to GPU
    input_size = X_sme_torch.shape[1]  # Number of features
    hidden_size = 1024
    num_classes = len(np.unique(y_sme))  # Number of unique class labels
    model = SimpleClassifier(input_size, hidden_size, num_classes)
    model = model.cuda()  # Move the model to GPU

    # Move the data to GPU
    X_train = X_train.cuda()
    y_train = y_train.cuda()
    X_test = X_test.cuda()
    y_test = y_test.cuda()

    # Define loss function and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=0.001)
    num_epochs = 200


    # Training loop
    for epoch in range(num_epochs):
        optimizer.zero_grad()
        outputs = model(X_train)
        loss = criterion(outputs, y_train)
        loss.backward()
        optimizer.step()


    # Evaluate the model on the test data
    model.eval()
    with torch.no_grad():
        test_outputs = model(X_test)
        _, predicted = torch.max(test_outputs, 1)
        accuracy = accuracy_score(y_test.cpu(), predicted.cpu())
        conf_mat = confusion_matrix(y_test.cpu(), predicted.cpu())
        print(f"Test Accuracy: {accuracy:.4f}")
        print(conf_mat)
        conf_mat_sum_nn.append(conf_mat)
    
    print('********************************************')
    test_acc.append(accuracy)
```
